```python
import pandas as pd

melb_df = pd.read_csv('data/melb_data_fe.csv')
melb_df.head()
display(melb_df.info())
```

```python
"""Задание 1.1. Преобразуйте столбец Date в формат datetime и выделите квартал (quarter) продажи объектов недвижимости.
Найдите второй по популярности квартал продажи. В качестве ответа запишите число объектов, проданных в этом квартале."""
melb_df['Date'] = pd.to_datetime(melb_df['Date'])
melb_df['month'] = melb_df['Date'].dt.month
melb_df['Quarter'] = (melb_df['month'] - 1) // 3 + 1 # Создаем столбец 'Quarter' для представления квартала
melb_df.head()
```

```python
def get_quarter(arg):
    if arg > 0 and arg < 4  : return 1
    if arg > 3 and arg < 7  : return 2
    if arg > 6 and arg < 10 : return 3
    if arg > 9 and arg < 13 : return 4
    else: return "error"

quarter_series = melb_df['month']

quarter = quarter_series.apply(get_quarter)
print(quarter)
```

```python
quarter_counts = melb_df['Quarter'].value_counts().reset_index()
#print(quarter_counts) # выводим Группировку DataFrame по 'Quarter' и считаем количество строк в каждой группе
quarter_counts.columns = ['Quarter', 'Count']
quarter_counts = quarter_counts.sort_values(by='Count')

# Выводим результат
print(quarter_counts)
```

Задание 1.2
Преобразуйте все столбцы, в которых меньше 150 уникальных значений, в тип данных category, исключив из преобразования столбцы Date, Rooms, Bedroom, Bathroom, Car.
В качестве ответа запишите результирующее количество столбцов, которые имеют тип данных category.

```python
cols_to_exclude = ['Date', 'Rooms', 'Bedroom', 'Bathroom', 'Car'] # список столбцов, которые мы не берём во внимание
max_unique_count = 150 # задаём максимальное число уникальных категорий
calc = 0
for col in melb_df.columns: # цикл по именам столбцов
    if melb_df[col].nunique() < max_unique_count and col not in cols_to_exclude: # проверяем условие
        melb_df[col] = melb_df[col].astype('category') # преобразуем тип столбца
for col in melb_df.columns: # цикл по именам столбцов
    if melb_df[col].dtype == 'category':
        calc = calc+1


print("calc = ", calc)
display(melb_df.info())
```

Задание 3.3
Какая риелторская компания (SellerG) имеет наименьшую общую выручку за период с 1 мая по 1 сентября (включительно) 2017 года?
Для ответа на этот вопрос рассчитайте сумму продаж (Price) каждой компании в заданный период.
Не забудьте перевести даты в формат datetime.

```python
#melb_df.head()
melb_df["Date"] = pd.to_datetime(melb_df["Date"])
#print(melb_df["Date"])
d1 = pd.to_datetime('2017-05-01')
d2 = pd.to_datetime('2017-09-02')
condition = (melb_df["Date"] > d1) & (melb_df["Date"] < d2)
SellerG_best = melb_df[condition].groupby(by = 'SellerG')['Price'].sum().sort_values(ascending=True)
print(SellerG_best)
```

Составьте сводную таблицу, которая показывает зависимость медианной площади (BuildingArea) здания от типа объекта недвижимости (Type) и количества жилых комнат в доме (Rooms). Для какой комбинации признаков площадь здания наибольшая?
В качестве ответа запишите эту комбинацию (тип здания, число комнат) через запятую, без пробелов.
