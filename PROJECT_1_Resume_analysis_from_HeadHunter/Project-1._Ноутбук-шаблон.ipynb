* Признак пола должен иметь 2 уникальных строковых значения: 'М' - мужчина, 'Ж' - женщина.

Выполните преобразование, ответьте на контрольные вопросы и удалите признак **"Пол, возраст"** из таблицы.

Гарантируется, что структура одинакова для всех строк в таблице. Вы можете воспользоваться этим.*


```python
#ваш код здесь

#переменные для выделения пола
new_column_name_1 = 'Пол'
base_column_list = ["Мужчина", "Женщина"]
new_column_list = ["М", "Ж", "other"]

##переменные для выделения возраста
new_column_name_2 = 'Возраст'

#общие переменные для всего задания
base_column_name = 'Пол, возраст'
error = "other"

df = hhdb # переназначаем дата фрейм для возможности использовать класс в будующем

if base_column_name in hhdb.columns: #Проверяем совершённость операций над колонкой
        
    class creating_feature():

        def __init__(self, new_column_name):
            df[new_column_name] = pd.Series()

        # Поиск и создание новых значений по спискам
        def def_feature_1(self, base_column, base_column_list, new_column_list):
            for i, ii  in  zip(base_column_list, new_column_list):
                if i in base_column:
                    return ii
            return "error"
        
        # Выделение значения между первой и второй запятой
        def def_feature_2(self, base_column_str): 
            # Разделение строки по запятой
            split_str = base_column_str.split(',')
            # Выделение значения между первой и второй запятой
            if len(split_str) >= 3:
                str_between_commas = split_str[1].strip()
                #выделяем числовую часть
                str_age = str_between_commas.split(' ')[0].strip()
                return str_age
            else: return "error"

    #Создание колонки 'Пол'
    new_column_1 = creating_feature(new_column_name_1)
    df[new_column_name_1] = df[base_column_name].apply(lambda x: new_column_1.def_feature_1
    (x, base_column_list, new_column_list))

    #Создание колонки 'Возраст'
    new_column_2 = creating_feature(new_column_name_2)
    df[new_column_name_2] = df[base_column_name].apply(lambda x: new_column_2.def_feature_2(x))

    hhdb = df #применяем изменения к нашей базе

    hhdb = hhdb.drop('Пол, возраст', axis=1) #Удалим ненужную колонку

else: print("столбец <",  base_column_name, "> уже удалён \n")

print("Проверяем данные \n", hhdb['Пол'].value_counts(), "\n")
print("Задание 3.2; Сколько процентов женских резюме представлено?")
print("Процент женских резюме =", round(len(hhdb[hhdb['Пол']=='Ж']) / hhdb.shape[0] * 100, 2), "\n")

print("Задание 3.2; Чему равен средний возраст соискателей?")
hhdb['Возраст'] = hhdb['Возраст'].astype(int) #преобразовываем данные
print("cредний возраст соискателей =", round(hhdb['Возраст'].mean(), 1))
```

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.

```python

#ваш код здесь

if "Опыт работы (месяц)" not in hhdb.columns: # Проверяем выполняли ли уже данный блок

    import re
    experience = pd.DataFrame()

    # Паттерны для поиска опыта работы в тексте
    years_pattern = r'(\d+) (?:лет|года|год)'
    months_pattern = r'(\d+) (?:месяцев|месяца|месяц)'

    # Применение паттернов к DataFrame
    experience['лет'] = hhdb['Опыт работы'].apply(lambda x: re.search(years_pattern, ' '.join(str(x).split()[:4])).group(1) 
                                                if re.search(years_pattern, ' '.join(str(x).split()[:4]))
                                                else None)
    experience['месяцев'] = hhdb['Опыт работы'].apply(lambda x: re.search(months_pattern, ' '.join(str(x).split()[:6])).group(1) 
                                                    if re.search(months_pattern, ' '.join(str(x).split()[:6]))
                                                    else None)
    experience = experience.astype(float)

    hhdb.drop('Опыт работы', axis=1, inplace=True) # Удаляем колонку в соответствии с условием
    # Вычисляем значение месяцев учитывая что данные могут быть NaN
    hhdb["Опыт работы (месяц)"] = (experience['лет'].fillna(0) * 12 + experience['месяцев'].fillna(0)).astype(int)
    del experience

    
else: print ("Колонка [Опыт работы (месяц)], содержащая колличество месяцев трудового стажа уже создана \n")
    
print(" Задание 3.3 Медеана по общему колличеству месяцев работы \n",
        hhdb["Опыт работы (месяц)"].loc[hhdb["Опыт работы (месяц)"] != 0].median())
```

Создадим отдельные признаки **"Город"**, **"Готовность к переезду"**, **"Готовность к командировкам"**. При этом важно учесть:

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
#ваш код здесь

if "Город, переезд, командировки" in hhdb.columns: # Проверяем удалена ли колонка

    million_cities = ["Москва", "Санкт-Петербург", 'Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск',
                        'Омск', 'Самара','Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']
    # Сроздаём Дата Фрейм для работы над заданием
    town_df = pd.DataFrame({"str": hhdb["Город, переезд, командировки"]})

    # Создаём и заполняем колонку город
    town_df['Город'] = town_df['str'].apply(lambda x: x.split()[0] if x.split()[0] in million_cities else 'другие')
    hhdb['Город'] = town_df['Город']

    # Задание 3.4 Сколько процентов соискателей живут в Санкт-Петербурге? Ответ округлите до целого.
    # Подсчет количества каждого уникального значения в столбце 'Город'
    city_counts = town_df['Город'].value_counts()
    spb_count = city_counts.get("Санкт-Петербург", 0) # Получение количества значений "Санкт-Петербург"
    total_count = len(town_df) # Получение общего количества значений в столбце 'Город'
    spb_percentage = (spb_count / total_count) * 100 # Расчет процента значений "Санкт-Петербург"


    # Определение объекта re.Pattern с использованием регулярного выражения
    pattern = re.compile(r',\s*([^,]+перее[^\d,]+?)[\s,;)(]+')
    # Применение шаблона к каждой строке и получение найденной строки
    town_df['relocate'] = town_df['str'].apply(lambda x: pattern.search(x).group(1) if pattern.search(x) else None)

    # Получаем уникальные значения колонки 'relocate'
    unique_relocate = town_df['relocate'].unique()
    #print("уникальные значения колонки 'relocate': \n", unique_relocate, "\n")

    # ОФункция замены на булевые значения
    def replace_condition(row):
        if row['relocate'] in unique_relocate[:2]: return False 
        elif row['relocate'] in unique_relocate[2:]: return True
        else: return "error"

    # Применение условия к каждой строке на булевые значения
    town_df['relocate'] = town_df.apply(lambda row: replace_condition(row), axis=1)
    # проверяем на наличие ошибок
    #print("уникальные значения колонки 'relocate': \n", town_df['relocate'].unique(), "\n")

    # Определение объекта re.Pattern с использованием регулярного выражения
    pattern = re.compile(r'(?<=,)\s*([^,]+ком[^\d,]+?)(?=[\s,;)(]|$)')
    # Применение шаблона к каждой строке и получение найденной строки
    town_df['business trip'] = town_df['str'].apply(lambda x: pattern.search(x).group(1) if pattern.search(x) else None)

    unique_business_trip = town_df['business trip'].unique()
    #print("список уникальных значений колонки business_trip \n", pd.Series(unique_business_trip))

    # Функция замены на булевые значения
    def replace_trip(row):
        if row['business trip'] in unique_business_trip[[1, 3, 4, 5, 7, 8]]: return True
        elif row['business trip'] in unique_business_trip[[0, 2, 6]]: return False
        else: return "error"

    # Применение условия к каждой строке и замена на булевые значения
    town_df['business trip'] = town_df.apply(lambda row: replace_trip(row), axis=1)
    # проверяем на наличие ошибок
    #print("уникальные значения колонки 'business trip': \n", town_df['business trip'].unique(), "\n") 

    # Определяем одновременую готовность к командировкм  и к переезду
    town_df['relocate_business trip'] = town_df['business trip'] & town_df['relocate']
    # Рассчитываем процентное соотношение только для True
    percentage_true = (town_df['relocate_business trip'] == True).sum() / len(town_df) * 100

    hhdb["Готовность к командировкам"]=town_df['business trip']
    hhdb["Готовность к переезду"]=town_df['relocate']

    del town_df # Удаляем прмежуточный Дата Фрейм

    hhdb.drop("Город, переезд, командировки", axis=1, inplace=True) # Удаляем колонку в соответствии с условием

else: print("[Город, переезд, командировки], уже удалена \n")

print("Процент соискателей живущих в  Санкт-Петербурге: \n", spb_percentage, "\n")   
print("одновременная готовность к командировкм  и к переезду \n", percentage_true)
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
#ваш код здесь

#функция для извлечения сущностей разделённых запятыми и создания столбцов для оных
def df_true_false(df, column):
    try:
        unique_str = [] # Cписок для уникальных фраз
        # Обходим каждую строку в столбце
        for row in df[column]:
            phrases = ','.join(map(lambda x: x.strip(), row.split(',')))
            phrases = phrases.split(',') # Разбиваем строку по запятой
            unique_str.extend(set(phrases))    # Добавляем  фразы в список

        # Преобразуем список в множество, уникальные значения, сортировка
        unique_str = sorted(list(set(unique_str)))
        # Вывод уникальных фраз для проверки
        #print("список уникальных сущностей столбца = ", column.name,": \n", list(unique_str), "\n")

        for string in unique_str:
            df[string] = df[column].apply(lambda x: string in x.split(', '))
        
        df.drop(column, axis=1, inplace=True) # Удаляем колонку в соответствии с условием

    except KeyError as e:
        print("Задание уже выполнено колонки удалены")

df_true_false(hhdb, 'Занятость')
df_true_false(hhdb, "График")

# Подсчет количества строк, в которых оба столбца содержат True
def true_true(column1, column2):
    count_true_rows = (column1 & column2).sum()
    return count_true_rows

task_3_5 = true_true(hhdb["волонтерство"], hhdb['проектная работа'])
print("task_3_5 = ", task_3_5)

task_3_6 = true_true(hhdb['вахтовый метод'], hhdb['гибкий график'])
print("task_3_6 = ", task_3_6)
```

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
#ваш код здесь

hhdb = hhdb.rename(columns={'Обновление резюме': 'data'}) #Приводим столбцы к одинаковым наименованиям

# грузим файл с котировками, создаём таблицу
exchangerates = pd.read_csv('data/ExchangeRates.csv')
if 'proportion' in exchangerates.columns:
    # переводим дату в Дата Фрейм
    exchangerates['date'] = pd.to_datetime(exchangerates['date'], format='%d/%m/%y', errors='coerce')
    # приводим всё к одной единице валюты
    exchangerates['close'] = exchangerates['close'] / exchangerates['proportion']
    # Удаляем данные о пропорции
    exchangerates = exchangerates.drop('proportion', axis=1)
    exchangerates['close'] = exchangerates['close'].astype(float)
else: print("столбец proportion уже удалён \n")

exchangerates = exchangerates.rename(columns={'date': 'data'}) #Приводим столбцы к одинаковым наименованиям
#print("Проверяем какие валюты у нас есть:\n ", exchangerates['currency'].unique())

# 1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. 
#   тот же формат привести признак "date" из таблицы с валютами.

hhdb["data"] = pd.to_datetime(hhdb['data'], format='%d.%m.%Y %H:%M', errors='coerce').dt.date
hhdb['data'] = pd.to_datetime(hhdb['data'], errors='coerce')

# 2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется.
#   Наименование валюты перевести в стандарт ISO согласно с таблицей выше.

if 'currency' not in hhdb.columns:
    hhdb['currency'] = hhdb['ЗП'].str.extract(r'(\D+)', expand=False)
    # Удаление начальных пробелов, если они есть
    hhdb['currency'] = hhdb['currency'].str.lstrip()
    # Извлечение цифр из столбца 'ЗП'
    hhdb['ЗП'] = hhdb['ЗП'].str.extract(r'(\d+)', expand=False)
    # Преобразование в числа и заполнение NaN
    hhdb['ЗП'] = pd.to_numeric(hhdb['ЗП'], errors='coerce')
    # Строки, которые невозможно преобразовать, добавить в список error
    error_list = hhdb[hhdb['ЗП'].isna()]['ЗП'].tolist()
    #print("строки которые не удалось преобразовать", error_list)
    
    hhdb = hhdb.rename(columns={'Обновление резюме': 'data'})
else: print("столбец currency уже создан")

# Смотрим зарплаты в каких валютах представлены в резюме 
#print("валюты в резюме: \n", hhdb['currency'].unique())

#необходимо заменить ('грн.', 'сум', 'руб.', 'бел.руб.')
list_for_replacement = ['грн.', 'сум', 'бел.руб.', 'руб.'] # порядок важен
list_for_replace = ['UAH', 'UZS', 'BYN', 'RUB']  # порядок важен

#заменяем ('грн.', 'сум', 'руб.', 'бел.руб.') на ['UAN', 'UZS', 'RUB', 'BYN']
for original, replacement in zip(list_for_replacement, list_for_replace):
    hhdb['currency'] = hhdb['currency'].str.replace(original, replacement)
#print("Проверяем какие валюты у нас есть:\n ", hhdb['currency'].unique())

# Меняем местами столбцы
currency = hhdb['currency']
hhdb = hhdb.drop('currency', axis=1)
hhdb.insert(1, 'currency', currency)

# 3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты
# (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные
# о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить
# единицей 1 (курс рубля самого к себе)

if 'close' not in hhdb.columns:
    # Объединение данных по столбцам 'data' и 'currency'
    merged_df = pd.merge(hhdb, exchangerates[['data', 'currency', 'close']], left_on=['data', 'currency'],
                         right_on=['data', 'currency'], how='left')
    # Заполнение столбца 'close' в основном DataFrame
    hhdb['close'] = merged_df['close'] 

    # Заполнить пропущенные значения в колонке "close" значениями на 1
    hhdb['close'] = hhdb.apply(lambda row: 1 if row['ЗП'] != 0 and row['currency'] == 'RUB' else row['close'], axis=1)
else: print("колонка close уже создана")

# 4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию
# (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)"."""

hhdb['ЗП (руб)'] = hhdb["ЗП"] *  hhdb['close']

#Задание 3.6
print("Задание 3.6 = ", hhdb['ЗП (руб)'].median())
```

```python
# ваш код здесь
import matplotlib.pyplot as plt
import seaborn as sns

def hist_box_plot(data, column, additional_line_value = 1, xy_x = 0.40, log=False):
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(15, 4))
    data_orig = data[column]
    data_fig = data[column]
    if log:
        data_fig = np.log10(data_fig)

    m_value = data_fig.mode().mean() #используем для ответа на задание 4.1
    std_value = data_fig.std()

    histplot = sns.histplot(data_fig, ax=axes[0], label='Histogram')
    histplot.set_title(f'Распределение [{column}]')
    histplot.set_ylabel('Количество')
    histplot.axvline(m_value, color='k', lw=2)
    histplot.axvline(m_value + 3 * std_value, color='k', ls='--')
    histplot.axvline(m_value - 3 * std_value, color='k', ls='--')

    boxplot = sns.boxplot(data_fig, orient='h', ax=axes[1])
    boxplot.set_title(f'Ящик с усами для [{column}]')
    boxplot.axvline(additional_line_value, color='r', linestyle='--', label='Additional Line')

    outliers_upper = data[data_fig > m_value + 3 * std_value]
    outliers_lower = data[data_fig < m_value - 3 * std_value]
    min_value_upper = outliers_upper[column].min()
    max_value_lower = outliers_lower[column].max()

    legend_hist = f"для гистограмы используем \nметод трёх сигм:\n" \
                  f"выбросы слева менее {max_value_lower}\n" \
                  f"мода = {round(data_orig.mode().mean())}\n" \
                  f"выбросы справа более {round(min_value_upper)}\n"
    histplot.annotate(legend_hist, xy=(0.65, 0.95), xycoords='axes fraction', fontsize=8, ha='left', va='top')

    # Вычисляем параметры ящика с усами
    lower_quartile = np.percentile(data_orig, 25)  # Нижний квартиль
    upper_quartile = np.percentile(data_orig, 75)  # Верхний квартиль
    interquartile_range = upper_quartile - lower_quartile  # Межквартильный размах
    lower_whisker = np.min(data_orig[data_orig >= lower_quartile - 1.5 * interquartile_range])  # Нижний ус
    upper_whisker = np.max(data_orig[data_orig <= upper_quartile + 1.5 * interquartile_range])  # Верхний ус
    median = np.median(data_orig)  # Медиан

    legend_box = f"Параметры коробочной диаграммы для \n [{column}]:\n"\
                 f"Левая граница уса: {round(lower_whisker)}\n" \
                 f"Левая граница коробки: {round(lower_quartile)}\n" \
                 f"Медиана: {round(median)}\n" \
                 f"Правая граница коробки: {round(upper_quartile)}\n" \
                 f"Правая граница уса: {round(upper_whisker)}\n"
    boxplot.annotate(legend_box, xy=(xy_x, 0.95), xycoords='axes fraction', fontsize=8, ha='left', va='top')

hist_box_plot(hhdb, "Возраст", 75)
```

```python
import textwrap # для отображения текста на блоках

gist_cat = "Образование"
gist_values = 'ЗП (руб)'
max_zp = 1000000

def gist_cat_def(df, gist_cat, gist_values, max_zp=None):
    if max_zp is not None:
        df = df[df[gist_values] < max_zp]    
    # Сортируем DataFrame по значениям "ЗП (руб)"
    df_sorted = df[[gist_cat, gist_values]].copy().sort_values(by=gist_values)
    gist_name = "Медианная"
    series = df_sorted.groupby(gist_cat)[gist_values].median()
    fig, ax = plt.subplots()
    bars = ax.barh(series.index, series.values)  # Заменили bar на barh

    # Переносим слова в подписях
    labels = [textwrap.fill(label, width=10) for label in df_sorted[gist_cat]]

    # Убираем подписи по оси "y"
    ax.set_yticks([])

    # Добавление аннотаций к столбцам
    for i, (labels, value) in enumerate(zip(series.index, series.values)):
        bbox_props = dict(boxstyle="round,pad=0.3", fc="white", ec="black", lw=0.0, alpha=0.0)
        ax.annotate(textwrap.fill(labels, width=20),
                    (0, i),  # Размещаем текст в начале основания столбца
                    xytext=(5, 0),  # Сдвигаем текст вправо
                    textcoords='offset points',
                    ha='left', va='center', rotation=0, fontsize=12, bbox=bbox_props, color='white')

    # Добавление подписей
    plt.title(f"{gist_name} [{series.name}] группировка по [{series.index.name}]")
    plt.xlabel(f"{gist_name} [{series.name}]")
    plt.ylabel(series.index.name)

    # Отображаем
    plt.show()

gist_cat_def(hhdb, gist_cat, gist_values, max_zp)
```
